```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Fish.csv")
```

```python
print(list(dataset['Species'].unique()))
```

```python
df =  pd.DataFrame(dataset[dataset['Species'].isin(['Bream','Perch'])])
```

```python
df.index = range(len(df))
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
```

```python
sc = StandardScaler()
df.iloc[:,1:]=sc.fit_transform(df.iloc[:,1:])
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(df.iloc[:,1:].values,df['Species'].values,test_size = 0.15)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NB_Model = GaussianNB()
NB_Model.fit(xtrain,ytrain)
```

```python
print(NB_Model.score(xtrain,ytrain))
```

```python
ypred=NB_Model.predict(xtest)
print(ypred)
```

```python
cm = confusion_matrix(ytest,ypred)
```

```python
print(cm)
```

```python
print(classification_report(ytest,ypred))
```

```python
print("accuracy : ",(cm[0][0]+cm[1][1])/len(ytest))
```

```python
#f1-score eval the overall relability of model in identifing the positive class  the higher the scroe indicate the better performance and detection of true positive
```

```python
probs = NB_Model.predict_proba(xtest)
probs = probs[:,1]
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelBinarizer
```

```python
LB = LabelBinarizer()
ytest_b = LB.fit_transform(ytest)
fpr,tpr,_ = roc_curve(ytest_b,probs)
random_probs = [0 for _ in range(len(ytest))]

p_fpr,p_tpr,_ = roc_curve(ytest_b,random_probs)
auc_score = roc_auc_score(ytest,probs)
print(auc_score)
```

```python
plt.plot(p_fpr,p_tpr,linestyle = '--')
plt.plot(fpr,tpr,marker = '.',label = 'Naive Bayes (area=%0.2f)' % auc_score)
plt.xlabel('False positive Rate')
plt.ylabel('True positive Rate')
plt.title('ROC-AUC Curve')
plt.legend()
plt.show()
```
